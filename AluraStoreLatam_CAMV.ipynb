#Ventas respecto al tiempo

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
```

```python
for i in range(len(tiendas)):
  tiendas[i]['tienda'] = f'Tienda {i+1}'
```

```python
tiendas = pd.concat(tiendas)
```

```python
tiendas.sample(5)
```

```python
tiendas['Fecha_de_compra'] = pd.to_datetime(tiendas['Fecha_de_compra'], format = '%d/%m/%Y')
tiendas['Fecha_de_compra'] = tiendas['Fecha_de_compra'].dt.strftime('%Y-%m')
```

```python
tiendas['Fecha_de_compra'] = pd.to_datetime(tiendas['Fecha_de_compra'], format = '%Y-%m')
```

```python
tiendas.info()
```

```python
tiendas = tiendas.sort_values('Fecha_de_compra')
```

```python
tendencia_mensual = tiendas.groupby([tiendas['Fecha_de_compra'].dt.to_period('M'), 'tienda'])['Precio'].sum().reset_index()
```

```python
tendencia_mensual
```

```python
tendencia_mensual['Fecha_de_compra'] = tendencia_mensual['Fecha_de_compra'].dt.to_timestamp()
```

##Gráfica

```python
from matplotlib import figure
plt.figure(figsize=(12, 6))
sns.lineplot(data=tendencia_mensual, x='Fecha_de_compra', y='Precio',
              hue='tienda', style='tienda', marker='o'
)

plt.title('Ventas Mensuales por Tienda: Comparativa de Crecimiento', fontsize=15)
plt.xlabel('Fecha de compra por trimestre')
plt.ylabel('Ventas')
plt.legend(title='Tiendas', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()

plt.savefig('Ventas_Mensuales_Por_Tiempo.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```
